```python
import matplotlib.pyplot as plt

categories = ['Category 1', 'Category 2', 'Category 3',
'Category 4']
values = [35, 20, 15, 30]

plt.pie(values, labels=categories, autopct='%1.1f%%')

plt.title('Pie Chart Example')

plt.show()
```
