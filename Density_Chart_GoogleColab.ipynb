```python
import matplotlib.pyplot as plt
import numpy as np

data = np.random.randn(1000)

plt.hist(data, density=True, bins=30, alpha=0.5)

plt.xlabel('Value')
plt.ylabel('Density')
plt.title('Density Plot')

plt.show()
```
